```python
def remove_high_missing_columns(df, threshold=0.3):

    # Calculate the missing value percentage for each column
    missing_percentage = df.isnull().mean()

    # Filter out columns exceeding the threshold
    columns_to_keep = missing_percentage[missing_percentage <= threshold].index

    # Return a DataFrame with the remaining columns
    return df[columns_to_keep]


# Removing columns with more than 30% missing values
dfprev = remove_high_missing_columns(dfprev, threshold=0.3)
dfprev.dropna(inplace=True)
```

```python
# Function to create synthetic Customer Master Data from Retail Sales Order dataset
def create_synthetic_customer_master(retail_df):
    """
    Creates a Customer Master Dataset with credit and AR-specific fields needed by CAs
    for accounts receivable management and credit assessment.
    """
    # Aggregate basic customer data
    customer_master = retail_df.groupby('Purchase Address').agg(
        First_Order_Date=('Order Date', 'min'),
        Last_Order_Date=('Order Date', 'max'),
        Total_Orders=('Order ID', 'nunique'),
        Total_Quantity=('Quantity Ordered', 'sum'),
        Avg_Order_Value=('turnover', 'mean'),
        Total_Turnover=('turnover', 'sum'),
        Total_Margin=('margin', 'sum')
    ).reset_index()
    
    customer_master['Last_Order_Date'] = pd.to_datetime(customer_master['Last_Order_Date'])
    customer_master['First_Order_Date'] = pd.to_datetime(customer_master['First_Order_Date'])


    # Calculate order frequency
    customer_master['Days_Customer'] = (customer_master['Last_Order_Date'] - customer_master['First_Order_Date']).dt.days
    customer_master['Days_Customer'] = customer_master['Days_Customer'].replace(0, 1)  # Avoid division by zero
    customer_master['Avg_Order_Frequency'] = (customer_master['Total_Orders'] / customer_master['Days_Customer']).round(2)
    
    # Add synthetic credit limit (based on turnover)
    np.random.seed(42)
    customer_master['Credit_Limit'] = (customer_master['Total_Turnover'] * np.random.uniform(1.5, 3.0, len(customer_master))).round(2)
    
    # Add synthetic credit score (random 300-850)
    customer_master['Credit_Score'] = np.random.randint(300, 851, len(customer_master))
    
    # Add synthetic credit status based on credit score
    conditions = [
        customer_master['Credit_Score'] >= 700,
        (customer_master['Credit_Score'] >= 600) & (customer_master['Credit_Score'] < 700),
        customer_master['Credit_Score'] < 600
    ]
    choices = ['Approved', 'Review', 'Rejected']
    customer_master['Credit_Status'] = np.select(conditions, choices, default='Review')
    
    # Add customer ID
    customer_master['Customer_ID'] = customer_master['Purchase Address'].apply(lambda x: abs(hash(x)) % 100000)
    
    # Add AR-specific fields for credit assessment
    customer_master['Default_Payment_Terms'] = np.random.choice(['Net 30', 'Net 45', 'Net 60'], len(customer_master))
    customer_master['Outstanding_Balance'] = (np.random.uniform(0, 0.3, len(customer_master)) * customer_master['Total_Turnover']).round(2)
    customer_master['Days_Outstanding'] = np.random.randint(0, 90, len(customer_master))
    customer_master['Contact_Name'] = ['Contact ' + str(i) for i in range(len(customer_master))]
    customer_master['Contact_Email'] = customer_master['Contact_Name'].str.replace(' ', '.').str.lower() + '@example.com'
    customer_master['Contact_Phone'] = ['(555) ' + str(np.random.randint(100, 999)) + '-' + str(np.random.randint(1000, 9999)) for _ in range(len(customer_master))]
    
    # Add credit approval date and next review date
    current_date = pd.Timestamp('2025-04-25')  # Using current date from query
    customer_master['Credit_Approval_Date'] = current_date - pd.to_timedelta(np.random.randint(30, 365), unit='D')
    customer_master['Credit_Review_Date'] = customer_master['Credit_Approval_Date'] + pd.Timedelta(days=365)
    
    
    # Calculate aging buckets - critical for AR analysis
    aging_conditions = [
        customer_master['Days_Outstanding'] == 0,
        customer_master['Days_Outstanding'] <= 30,
        customer_master['Days_Outstanding'] <= 60,
        customer_master['Days_Outstanding'] <= 90
    ]
    aging_choices = ['Current', '1-30 Days', '31-60 Days', '61-90 Days']
    customer_master['Aging_Bucket'] = np.select(aging_conditions, aging_choices, default='90+ Days')
    # Add available credit calculation
    customer_master['Available_Credit'] = (customer_master['Credit_Limit'] - customer_master['Outstanding_Balance']).round(2)
    
    return customer_master
```

```python
customer_master = create_synthetic_customer_master(dfsales)
```

```python
customer_master.to_csv('customer_master.csv', index=False)
print("Completed creating customer master data and saved to 'customer_master.csv'")
```

```python
def create_synthetic_sales_order(retail_df, loan_df):
    """
    Creates a comprehensive Sales Order dataset with AR-specific fields
    for accounts receivable tracking, with proper sequencing to avoid KeyErrors.
    """
    # Copy retail data to avoid modifying original
    sales_order = retail_df.copy()
    
    # Ensure Order Date is datetime format
    if sales_order['Order Date'].dtype != 'datetime64[ns]':
        sales_order['Order Date'] = pd.to_datetime(sales_order['Order Date'])
    
    # Add Customer ID based on address hash
    sales_order['Customer_ID'] = sales_order['Purchase Address'].apply(lambda x: abs(hash(x)) % 100000)
    
    # Add payment terms and order status - critical for AR
    np.random.seed(42)
    sales_order['Payment_Terms'] = np.random.choice(['Net 30', 'Net 45', 'Net 60'], len(sales_order))
    sales_order['Order_Status'] = np.random.choice(['Pending', 'Approved', 'Shipped', 'Delivered', 'Cancelled'], len(sales_order))
    
    # Add document tracking numbers
    sales_order['Invoice_Number'] = 'INV-' + (10000 + sales_order.index).astype(str)
    sales_order['PO_Number'] = 'PO-' + (20000 + sales_order.index).astype(str)
    
    # Map loan application status to credit check status
    unique_customers = sales_order['Customer_ID'].unique()
    unique_loans = loan_df['SK_ID_CURR'].unique()
    
    np.random.seed(42)
    mapping_size = min(len(unique_customers), len(unique_loans))
    customer_to_loan = dict(zip(unique_customers[:mapping_size], unique_loans[:mapping_size]))
    
    def get_credit_check_status(cust_id):
        loan_id = customer_to_loan.get(cust_id, None)
        if loan_id is None:
            return 'No Loan'
        
        # Safe way to check loan status
        loan_rows = loan_df[loan_df['SK_ID_CURR'] == loan_id]
        if loan_rows.empty:
            return 'No Loan'
            
        loan_status = loan_rows['NAME_CONTRACT_STATUS'].iloc[0]
        if loan_status == 'Approved':
            return 'Credit Approved'
        elif loan_status == 'Refused':
            return 'Credit Refused'
        else:
            return 'Under Review'
    
    # Apply credit check status
    sales_order['Credit_Check_Status'] = sales_order['Customer_ID'].apply(get_credit_check_status)
    
    # Calculate due dates based on payment terms
    def calculate_due_date(row):
        days_map = {'Net 30': 30, 'Net 45': 45, 'Net 60': 60}
        return row['Order Date'] + timedelta(days=days_map.get(row['Payment_Terms'], 30))
    
    sales_order['Due_Date'] = sales_order.apply(calculate_due_date, axis=1)
    
    # CREATE PAYMENT STATUS BEFORE REFERENCING IT - this prevents the KeyError
    payment_status_options = ['Paid', 'Unpaid', 'Partial', 'Overdue']
    payment_probs = [0.6, 0.2, 0.1, 0.1]
    sales_order['Payment_Status'] = np.random.choice(payment_status_options, len(sales_order), p=payment_probs)
    
    # Now we can safely reference Payment_Status
    sales_order['Days_Outstanding'] = 0
    unpaid_mask = sales_order['Payment_Status'].isin(['Unpaid', 'Overdue', 'Partial'])
    if unpaid_mask.sum() > 0:
        sales_order.loc[unpaid_mask, 'Days_Outstanding'] = np.random.randint(1, 90, unpaid_mask.sum())
    
    # Add remaining balance information
    sales_order['Remaining_Balance'] = 0.0
    sales_order.loc[sales_order['Payment_Status'] == 'Unpaid', 'Remaining_Balance'] = sales_order.loc[sales_order['Payment_Status'] == 'Unpaid', 'turnover']
    sales_order.loc[sales_order['Payment_Status'] == 'Overdue', 'Remaining_Balance'] = sales_order.loc[sales_order['Payment_Status'] == 'Overdue', 'turnover']
    
    # Handle partial payments carefully to avoid broadcasting errors
    partial_mask = sales_order['Payment_Status'] == 'Partial'
    if partial_mask.sum() > 0:
        partial_df = sales_order.loc[partial_mask]
        random_values = np.random.uniform(0.1, 0.9, len(partial_df))
        sales_order.loc[partial_mask, 'Remaining_Balance'] = partial_df['turnover'].values * random_values
    
    # Add aging buckets - essential for AR reporting
    aging_conditions = [
        sales_order['Days_Outstanding'] == 0,
        sales_order['Days_Outstanding'] <= 30,
        sales_order['Days_Outstanding'] <= 60,
        sales_order['Days_Outstanding'] <= 90
    ]
    aging_choices = ['Current', '1-30 Days', '31-60 Days', '61-90 Days']
    sales_order['Aging_Bucket'] = np.select(aging_conditions, aging_choices, default='90+ Days')
    
    # Add sales representative information
    sales_reps = ['John Doe', 'Jane Smith', 'Robert Johnson', 'Maria Garcia', 'David Kim']
    sales_order['Sales_Representative'] = np.random.choice(sales_reps, len(sales_order))
    
    # Add order approval details
    current_date = pd.Timestamp('2025-04-26')  # Use current date from query
    approvers = ['Susan Lee', 'Michael Brown', 'Jennifer Wilson', 'Thomas Moore']
    sales_order['Approved_By'] = np.where(
        sales_order['Order_Status'].isin(['Approved', 'Shipped', 'Delivered']),
        np.random.choice(approvers, len(sales_order)),
        None
    )
    
    # Add approval dates for approved orders
    approval_mask = sales_order['Approved_By'].notna()
    if approval_mask.sum() > 0:
        order_dates = sales_order.loc[approval_mask, 'Order Date']
        sales_order.loc[approval_mask, 'Approval_Date'] = order_dates + pd.to_timedelta(np.random.randint(0, 3, size=approval_mask.sum()), unit='D')
    
    # Add payment tracking information
    paid_mask = sales_order['Payment_Status'] == 'Paid'
    if paid_mask.sum() > 0:
        sales_order.loc[paid_mask, 'Payment_Amount'] = sales_order.loc[paid_mask, 'turnover']
        due_dates = sales_order.loc[paid_mask, 'Due_Date']
        sales_order.loc[paid_mask, 'Payment_Date'] = due_dates - pd.to_timedelta(np.random.randint(0, 15, size=paid_mask.sum()), unit='D')
    
    # Add partial payment details
    if partial_mask.sum() > 0:
        partial_df = sales_order.loc[partial_mask]
        remaining_ratio = partial_df['Remaining_Balance'] / partial_df['turnover']
        payment_ratio = 1 - remaining_ratio
        sales_order.loc[partial_mask, 'Payment_Amount'] = (partial_df['turnover'] * payment_ratio).values
        due_dates = partial_df['Due_Date']
        sales_order.loc[partial_mask, 'Payment_Date'] = due_dates - pd.to_timedelta(np.random.randint(0, 15, size=len(partial_df)), unit='D')
    
    return sales_order
```
